5 dimensions, using measurements defined by the pentagon geometry, complex phases and 4 rounds of measurements for Alice

```python
import numpy as np
from itertools import product
from itertools import permutations

# Set dimension
d = 5  # working in a 5-level system

# Useful functions for ket and ket ⊗ ket
def Ket(i):
    """Create a basis vector |i⟩ (0-indexed)"""
    return np.eye(1, d, i).flatten()  # More efficient than UnitVector

def ket(i, j):
    """Create a product state |i⟩ ⊗ |j⟩"""
    return np.kron(Ket(i), Ket(j))

def OuterProduct(psi, phi):
    """Create outer product |ψ⟩⟨φ|"""
    return np.outer(psi, phi.conj())

# Post measurement States
# x=1
Psi11 = 1/np.sqrt(2)*(Ket(0) + Ket(1))
Psi21 = 1/np.sqrt(2)*(Ket(0) - Ket(1))
Psi31 = 1/np.sqrt(2)*(Ket(1) + Ket(2))
Psi41 = 1/np.sqrt(2)*(Ket(1) - Ket(2))
Psi51 = 1/np.sqrt(2)*(Ket(2) + Ket(3))
Psi61 = 1/np.sqrt(2)*(Ket(2) - Ket(3))
Psi71 = 1/np.sqrt(2)*(Ket(3) + Ket(4))
Psi81 = 1/np.sqrt(2)*(Ket(3) - Ket(4))
Psi91 = 1/np.sqrt(2)*(Ket(0) + Ket(4))
Psi101 = 1/np.sqrt(2)*(Ket(0) - Ket(4))

# x=2
Psi12 = 1/np.sqrt(2)*(Ket(0) + 1j*Ket(1))
Psi22 = 1/np.sqrt(2)*(Ket(0) - 1j*Ket(1))
Psi32 = 1/np.sqrt(2)*(Ket(1) + 1j*Ket(2))
Psi42 = 1/np.sqrt(2)*(Ket(1) - 1j*Ket(2))
Psi52 = 1/np.sqrt(2)*(Ket(2) + 1j*Ket(3))
Psi62 = 1/np.sqrt(2)*(Ket(2) - 1j*Ket(3))
Psi72 = 1/np.sqrt(2)*(Ket(3) + 1j*Ket(4))
Psi82 = 1/np.sqrt(2)*(Ket(3) - 1j*Ket(4))
Psi92 = 1/np.sqrt(2)*(Ket(0) + 1j*Ket(4))
Psi102 = 1/np.sqrt(2)*(Ket(0) - 1j*Ket(4))


# x=3
Psi13 = 1/np.sqrt(2)*(Ket(0) + Ket(2))
Psi23 = 1/np.sqrt(2)*(Ket(0) - Ket(2))
Psi33 = 1/np.sqrt(2)*(Ket(2) + Ket(4))
Psi43 = 1/np.sqrt(2)*(Ket(2) - Ket(4))
Psi53 = 1/np.sqrt(2)*(Ket(1) + Ket(4))
Psi63 = 1/np.sqrt(2)*(Ket(1) - Ket(4))
Psi73 = 1/np.sqrt(2)*(Ket(1) + Ket(3))
Psi83 = 1/np.sqrt(2)*(Ket(1) - Ket(3))
Psi93 = 1/np.sqrt(2)*(Ket(0) + Ket(3))
Psi103 = 1/np.sqrt(2)*(Ket(0) - Ket(3))

# x=4
Psi14 = 1/np.sqrt(2)*(Ket(0) + 1j*Ket(2))
Psi24 = 1/np.sqrt(2)*(Ket(0) - 1j*Ket(2))
Psi34 = 1/np.sqrt(2)*(Ket(2) + 1j*Ket(4))
Psi44 = 1/np.sqrt(2)*(Ket(2) - 1j*Ket(4))
Psi54 = 1/np.sqrt(2)*(Ket(1) + 1j*Ket(4))
Psi64 = 1/np.sqrt(2)*(Ket(1) - 1j*Ket(4))
Psi74 = 1/np.sqrt(2)*(Ket(1) + 1j*Ket(3))
Psi84 = 1/np.sqrt(2)*(Ket(1) - 1j*Ket(3))
Psi94 = 1/np.sqrt(2)*(Ket(0) + 1j*Ket(3))
Psi104 = 1/np.sqrt(2)*(Ket(0) - 1j*Ket(3))
```

```python
def density_matrix(psi):
    return OuterProduct(psi, psi)
```

```python
computational_basis = [
    ("|0>", Ket(0)),
    ("|1>", Ket(1)),
    ("|2>", Ket(2)),
    ("|3>", Ket(3)),
    ("|4>", Ket(4))
]

all_povm_orders = list(permutations(computational_basis))
```

```python
def is_antidistinguishable(vectors, povm_kets, tol=1e-10):
    """
    Conditions:

    1. <psi_i | phi_i> = 0 for i = 1,2,3,4
    2. <psi_i | phi_5> = 0 for AT LEAST ONE i
    """

    # First condition:
    # <psi_i | phi_i> = 0
    for psi, phi in zip(vectors, povm_kets[:4]):

        overlap = np.vdot(phi, psi)

        if abs(overlap) > tol:
            return False

    # Second condition:
    # <psi_i | phi_5> = 0 for at least one i
    phi5 = povm_kets[4]

    phi5_orthogonal = False

    for psi in vectors:

        overlap5 = np.vdot(phi5, psi)

        if abs(overlap5) < tol:
            phi5_orthogonal = True
            break

    return phi5_orthogonal
```

```python
remaining_combinations = list(product(
    states["x1"],
    states["x2"],
    states["x3"],
    states["x4"]
))

for povm in all_povm_orders:

    povm_labels = [n for n, _ in povm]
    povm_vectors = [v for _, v in povm]

    valid_combinations = []
    new_remaining = []

    for combo in remaining_combinations:

        names = [n for n, _ in combo]
        vectors = [v for _, v in combo]

        if is_antidistinguishable(vectors, povm_vectors):

            valid_combinations.append(names)

        else:
            new_remaining.append(combo)

    # Update remaining combinations
    remaining_combinations = new_remaining

    print("\nPOVM ordering:", povm_labels)

    print("Number of valid combinations:",
          len(valid_combinations))

    #for c in valid_combinations:
        #print(c)

    print("Remaining combinations:",
          len(remaining_combinations))
```
