SDP to find basis that antidistinguish a specific set of states

```python
# Example: 4 pure qubit states
import numpy as np
import cvxpy as cp

psi1 = np.array([1, 1, 0]) / np.sqrt(2)
psi2 = np.array([1, 1j, 0]) / np.sqrt(2)
psi3 = np.array([0, 1, 1]) / np.sqrt(2)
psi4 = np.array([0, 1, 1j]) / np.sqrt(2)

rho = {}

rho[0] = np.outer(np.conj(psi1), psi1)
rho[1] = np.outer(np.conj(psi2), psi2)
rho[2] = np.outer(np.conj(psi3), psi3)
rho[3] = np.outer(np.conj(psi4), psi4)

n = len(rho)
d = rho[0].shape[0]

# --- CVXPY SDP ---
M = [cp.Variable((d, d), hermitian=True) for _ in range(n)]

# Objective: minimize sum Tr(rho_i M_i)
obj = 0
for i in range(n):
    obj += cp.trace(rho[i] @ M[i])

objective = cp.Minimize(cp.real(obj))

# Constraints
constraints = []

# Positivity
for i in range(n):
    constraints.append(M[i] >> 0)

# Completeness
sumM = 0
for i in range(n):
    sumM += M[i]

constraints.append(sumM == np.eye(d))

# Solve
prob = cp.Problem(objective, constraints)
prob.solve()

optval = prob.value
M_val = [Mi.value for Mi in M]

np.set_printoptions(
    precision=4,      # number of decimals
    suppress=True,    # avoid scientific notation
)

print(f"optval = {optval:.6e}\n")

for i, Mi in enumerate(M_val, start=1):
    print(f"M_{i} =")
    print(Mi)
    print()
```
